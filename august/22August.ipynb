```python
#Tuples

l = (1, 2, 3)
print(l)
```

```python
#access tuple items
print(l[2])
```

```python
#length of tuple
print(len(l))
```

```python
#Type
x = (1, ) # , is mandatory for 1 element in tuple
print(type(x))
print(type(l))
```

```python
diff = (1, 'a')
print(type(diff))
print(diff)
```

```python
l.append(4) #no insersion or deletion of elements
print(l)
```

```python
#converting to list and append element
lis = list(l)
lis.append(4)
tup = tuple(lis)
print(tup)
```

```python
tup1 = (1, 2, 3)
tup2 = ('a', 'b', 'c')
tup3 = tup1 + tup2
print(tup3)
```

```python
#tuple element removal
lst1 = list(tup1)
lst1.remove(1)
tup1 = tuple(lst1)
print(tup1)
```

```python
tup = (10, 20, 30, 40, 50)

repeat = tup * 2
print("Repeated tuple:", repeat)

# Functions and Methods
print("Max value:", max(tup))
print("Min value:", min(tup))
print("Sum:", sum(tup))
print("Count of 20:", tup.count(20))
print("Index of 40:", tup.index(40))
```

```python
x = (1, 'a', 'b', 2, 3, 4)

"""
Unpacking of tuple

y = x[0]
z = x[1]
m = x[2:] rest is stored in m as list
"""
(y, z, *m) = x

print(y)
print(z)
print(m)
```

```python
(y, *z, m) = x
print(y)
print(z)
print(m)
```

```python
#traversing within tuple
for i in x:
    print(i)
```
