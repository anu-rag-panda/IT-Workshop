```python
list = []

n = int(input("enter the number of elements: "))

for i in range(0, n):
  temp = input("enter the element: ")
  list.append(temp)

print(list)
search = input("enter the element to be searched: ")
if search in list:
  print("element found.")
  print("index of the element is", list.index(search))
else:
  print("element not found.")

```

```python
#nested list
nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

print(nested_list[0][0])  # Output: 1
print(nested_list[1][2])  # Output: 6
print(nested_list[2][1])  # Output: 8
```

```python
#list comprehension
squares = [x**2 for x in range(1, 6)]
print(squares)
```

```python
names = ["milo", "sarah", "bruno", "arpita", "rosa"]
#find the names having 'o' and create a list
names_with_o = [name for name in names if 'o' in name]
print(names_with_o)
```

```python
names = ["milo", "sarah", "bruno", "arpita", "rosa"]
#find the names not having 'o' and create a list
names_with_o = [name for name in names if 'o' not in name]
print(names_with_o)
```

```python
#find the even number upto hundred
even_numbers = [x for x in range(1, 101) if x%2==0]
print(even_numbers)
```
