```python
#factorial of a number using recursive function
n = int(input("enter a number to find factorial: "))

def fact(num):
  if num == 0:
    return 1
  return num * fact(num-1)

answer = fact(n)

print("The factorial of", n, "is", answer)
```

```python
#fibonacci series inside a range
rmin = int(input("enter the minimum range: "))
rmax = int(input("enter the maximum range: "))

def fibonacci(min, max):
    num1 = 0
    num2 = 1
    for i in range(min, max+1):
        ans = num1 + num2
        num1 = num2
        num2 = ans
        if ans >= min and ans <= max:
            print(ans, end=" ")

print(f"The fibonacci series between {rmin} and {rmax} is: ")
if rmin == 0:
    print(0, 1, end=" ")
    fibonacci(rmin, rmax)
elif (rmin == 1 and rmax >= 1):
    print(1, end=" ")
    fibonacci(rmin, rmax)
else:
    fibonacci(rmin, rmax)
```

```python
#write a pattern
#give input as example e, using ascii value
'''
A
A B
A B C
A B C D
A B C D E
'''

character = input("enter a character: ")
ch = ord(character.upper())

for i in range(ord("A"), ch + 1):
  for j in range(ord("A"), i + 1):
    print(chr(j), end=" ")
  print()
```

```python
#number palindrome or not
num = int(input("enter a number: "))

temp = num
reverse = 0
while temp > 0:
  remainder = temp % 10
  reverse = (reverse * 10) + remainder
  temp = temp // 10

if num == reverse:
  print(f"{num} is a palindrome number")
else:
  print(f"{num} is not a palindrome number")
```

```python
# String palindrome check
text = input("Enter a string: ")
reversed_text = text[::-1]
print(reversed_text)
if text == reversed_text:
  print(f"'{text}' is a palindrome")
else:
  print(f"'{text}' is not a palindrome")
```
