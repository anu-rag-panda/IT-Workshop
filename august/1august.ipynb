star pattern


```
*
* *
* * *
```



````python
r = int(input("enter the number of rows: "))
for i in range(1, r+1, 1):
  for j in range(1, i+1, 1):
    print("* ", end="")
  print()
````

inverted star pattern

````python
r = int(input("enter the number of rows: "))
for i in range(r+1, 1, -1):
  for j in range(1, i+1 -1):
    print("*", end=" ")
  print()
````

````python
r = int(input("enter the number of rows: "))
for i in range(1, r+1, 1):
  for j in range(1, r+1-i, 1):
    print(" ", end="")
  for k in range(1, i+1, 1):
    print("*", end="")
  print()
````

````python
'''
***
 **
  *
'''
r = int(input("enter the number of rows: "))
for i in range(1, r+1, 1):
  for j in range(1, i, 1):
    print(" ", end="")
  for k in range(1, r-i+2, 1):
    print("*", end="")
  print()
````

````python
#write a program using a for loop that calulate a exponential. your program should ask the user for a base and an exponantial (exp) and calulate base^exp
#base = 2 to 10
#exp = 2 to 5
#for e

# Function to calculate factorial
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)

# Function to calculate e^x using Taylor series
def power_of_e(x, n_terms):
  result = 0
  for i in range(n_terms):
    result += (x**i) / factorial(i)
  return result

for i in range(2, 11, 1):
  print("the exp of", i, "is: ")
  for j in range(2, 6, 1):
    print(i, "^", j, "=", i**j)
  print()

print("the power of e is: ")
for k in range(2, 6, 1):
  # Using the power_of_e function with a reasonable number of terms (e.g., 10)
  print("e ^",k, "=", "{:.2f}".format(power_of_e(k, 10)))
````
