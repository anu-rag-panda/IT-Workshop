List is basically a collection of elements

```python
#list
me = ["anurag", 18, 78.38]
print(me)
```

```python
print(me[1])
```

```python
print(me[-1])
```

```python
#sliceing
list = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(list[2:5])
```

```python
print(list[:4])
```

```python
print(list[2:])
```

```python
print(list[-4:-1]) #-1 is omitted
```

```python
#change the item value
list[1] = "avocado"
print(list)
```

```python
#printing list using for loop
for x in list:
  print(x)
```

```python
#search element in least using if
if "apple" in list:
  print('yes')
```

```python
print(len(list))
```

```python
list.append("grapes")
print(list)
```

```python
list.insert(2, "banana")
print(list)
```

```python
list.remove("banana")
print(list)
```

```python
list.pop()
print(list)
```

```python
mylist = list.copy()
print(mylist)
```

```python
list.reverse()
print(list)
```

```python
list1 = [1, 2, 3]
list2 = ["a", "b", "c"]
list3 = list1 + list2
print(list3)
```

```python
list.sort()
print(list)
```
