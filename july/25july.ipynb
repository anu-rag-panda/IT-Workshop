to convert celcius to farenheit

````python
celcius = float(input("enter the temperature in celcius: "))

farenheit = (celcius * 9/5) + 32

print("temperature in farenheit is: ", farenheit)
````

compound interest

````python
capital = int(input("enter the capital: "))
rate = float(input("enter the compund rate: "))
time = int(input("enter the time: "))

amount = capital * (1 + (rate/100)) ** time

print("the compound interest for capital " , capital, " and rate ", rate, " and time ", time, " is ", amount)
````

electricity bill

```
unit                price
first 100 unit      no charge
next 100 unit       rs 5 per unit
after 200 unit      rs 10 per unit
```

````python
unit = int(input("enter the unit: "))

if unit <= 100:
  bill = 0
elif unit <= 200:
  bill = (unit-100)*5
elif (unit > 200):
  bill = 500 + (unit-200)*10

print("the electricity bill is: ", bill)
````

find equilateral , isoscales, scalene triangle founder (equilateral -> a=b=c ; scalene -> a!=b!=c ; isoscales -> a=b!=c)

````python
side1 = int(input("enter the first side: "))
side2 = int(input("enter the second side: "))
side3 = int(input("enter the third side: "))

if (side1 == side2 == side3):
  print("Equilateral triangle.")
elif (side1 == side2 or side2 == side3 or side1 == side3):
  print("Isoscales triangle.")
else:
  print("Scalene triangle.")
````

input age , sex (M, F), number of days and display wages

```
age             sex          wage/day
>=18 and <30     M           700
                 F           750
>=30 and <=40    M           800
                 F           850
```

````python
age = int(input("enter the age: "))
sex = input("enter the sex (M/F): ")
days = int(input("enter the number of days: "))

if (sex == 'M' or sex == 'm'):
  if (age >= 18 and age < 30):
    wage = 700 * days
    print("the wage is: ", wage)
  elif (age >= 30 and age <= 40):
    wage = 800 * days
    print("the wage is: ", wage)
  else:
    print("enter appropiate age")
elif (sex == 'F' or sex == 'f'):
  if (age >= 18 and age < 30):
    wage = 750 * days
    print("the wage is: ", wage)
  elif (age >= 30 and age <= 40):
    wage = 850 * days
    print("the wage is: ", wage)
  else:
    print("enter appropiate age")
else:
  print("invalid input")
````
