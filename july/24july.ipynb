if elif else (grade card)

```python
marks = int(input("enter the marks: "))

if (marks<=100 and marks>=90):
  print("Grade O")
elif (marks<=89 and marks>=80):
  print("Grade E")
elif (marks<=79 and marks>=70):
  print("Grade A")
elif (marks<=69 and marks>=60):
  print("Grade B")
elif (marks<=59 and marks>=50):
  print("Grade C")
elif (marks<=49 and marks>=40):
  print("Grade D")
else:
  print("Grade F")

```

find the sum of even and odd number upto certain range

```python
n = int(input("enter the range: "))
odd=0
even=0
for i in range (1, n+1, 2):
  odd += i
print("sum of odd numbers is",odd)
for i in range (2, n+1, 2):
  even += i
print("sum of even numbers is",even)
```

```python
n = int(input("enter the range: "))
e = 0
o = 0
for i in range (1, n+1, 1):
    if (i%2 == 0):
        e += i
    else:
        o += i
print("even number addition is", e)
print("odd number addition is", o)
```
