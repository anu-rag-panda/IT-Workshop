check whether a year is leapyear or not

```python
y = int(input("enter a year: "))
if (y%100 == 0):
  if (y%400 == 0):
    print(y, "is a leapyear")
  else:
    print(y, "is not a leapyear")
elif (y%4 == 0):
  print(y, "is a leapyear")
else:
  print(y, "is not a leapyear")
```

binary, hexadecimal, octal number input

```python
a = 0b1010 #binary number after 0b
b = 0xE7 #hexadecimal number after 0x
c = 0o17 #octal number after 0o

print("0b1010 in decimal is", a)
print("0xE7 in decimal is", b)
print("0o17 in decimal is", c)
```

converting decimal to binary, octal, hexadecimal

```python
a = 10
b = bin(a)
print(a) #binary
b = oct(a)
print(b) #octal
b = hex(a)
print(b) #hexadecimal
```
