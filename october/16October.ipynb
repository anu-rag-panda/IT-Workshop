```python
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}

print(thisdict)
```

```python
#printing a value by key
x = thisdict["model"]
print(x)
```

```python
#.get() method
y = thisdict.get("model")
print(y)
```

```python
#loop through dictionary
for x in thisdict:
  print(x)
```

```python
#print all values
for x in thisdict:
  print(thisdict[x])
```

```python
#length of the dictionary
print(len(thisdict))
```

```python
#check if "model" is present in the dictionary
if "model" in thisdict:
  print("YES, 'model' is present in thisdict")
```

```python
#adding value
thisdict["color"] = "red"
print(thisdict)
```

```python
#removing item
thisdict.pop("color")
print(thisdict)
```

```python
#copy a dictionary
my_dict = thisdict.copy()
print(my_dict)
```

```python
#nested dictionary

myClass = {
    "anurag": {
        "age": 20,
        "roll": 199
    },
    "anirban": {
        "age": 19,
        "roll": 118,
    },
    "aritra": {
        "age": 18,
        "roll": 169
    }
}
```

```python
print(myClass)
```

```python
x = myClass["anurag"]["age"]

print(x)
```

```python
myClass = {
    "anurag": {
        "age": 20,
        "roll": 199,
        "nickname": "panda"
    },
    "anirban": {
        "age": 19,
        "roll": 118,
        "nickname":"anirban"
    },
    "aritra": {
        "age": 18,
        "roll": 169,
        "nickname": "aritraa"
    }
}

#print(myClass)

x = myClass["aritra"]["nickname"]
print(f"aritra is {x}")
print()

for i in myClass:
    print(i)
    for x in myClass[i]:
        print(myClass[i][x], end=" ")
    print()
    print()
```
